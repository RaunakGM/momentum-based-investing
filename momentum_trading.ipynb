```python
colls=['Stock Name','Stock Price','Feb-July Return','Aug-Jan Return','Average']
six_month_return = pd.DataFrame(columns=colls)
path = "D:/softwares/vscode files/stocks/"
for (file_name,n,p) in zip(os.listdir(path),Name,price):
   df = pd.read_csv(f'D:/softwares/vscode files/stocks/{file_name}',index_col='Date ',parse_dates=True)
   x=df.loc["2023-02-01",'ltp ']
   if (type(x)) == str:
      if ',' in x:
         x=x.replace(',','')
         x=float(x)
      else:
         x=float(x)
   else:
         x=float(x)
   y=df.loc["2022-08-01",'ltp ']
   if (type(y)) == str:
      if ',' in y:
         y=y.replace(',','')
         y=float(y)
      else:
         y=float(y)
   else:
         y=float(y)
   z=df.loc['2022-02-01','ltp ']
   if (type(z)) == str:
      if ',' in z:
         z=z.replace(',','')
         z=float(z)
      else:
         z=float(z)
   else:
         z=float(z)
   yzreturn = ((y-z)/z)*100
   xyreturn = ((x-y)/y)*100
   avg = (yzreturn+xyreturn)/2
   six_month_return=six_month_return.append(
       pd.Series(
       [
       n,
       p,
       yzreturn,
       xyreturn,
       avg
       ],
       index=colls),
       ignore_index=True
   )
six_month_return
```

```python
collls=['Stock Name','Stock Price','Feb-April Return','May-July Return','Aug-Oct Return','Nov-Jan Return','Average']
three_month_return = pd.DataFrame(columns=collls)
path = "D:/softwares/vscode files/stocks/"
for (file_name,n,p) in zip(os.listdir(path),Name,price):
   df = pd.read_csv(f'D:/softwares/vscode files/stocks/{file_name}',index_col='Date ',parse_dates=True)
   x=df.loc["2022-08-01",'ltp ']
   if (type(x)) == str:
      if ',' in x:
         x=x.replace(',','')
         x=float(x)
      else:
         x=float(x)
   else:
         x=float(x)
   y=df.loc["2022-11-01",'ltp ']
   if (type(y)) == str:
      if ',' in y:
         y=y.replace(',','')
         y=float(y)
      else:
         y=float(y)
   else:
         y=float(y)
   z=df.loc['2023-02-01','ltp ']
   if (type(z)) == str:
      if ',' in z:
         z=z.replace(',','')
         z=float(z)
      else:
         z=float(z)
   else:
         z=float(z)
   w=df.loc['2022-05-02','ltp ']
   if (type(w)) == str:
      if ',' in w:
         w=w.replace(',','')
         w=float(w)
      else:
         w=float(w)
   else:
         w=float(w)
   v=df.loc['2022-02-01','ltp ']
   if (type(v)) == str:
      if ',' in v:
         v=v.replace(',','')
         v=float(v)
      else:
         v=float(v)
   else:
         v=float(v)
   wvreturn = ((w-v)/v)*100
   xwreturn = ((x-w)/w)*100
   xyreturn = ((y-x)/x)*100
   yzreturn = ((z-y)/y)*100
   avg = (wvreturn+xwreturn+xyreturn+yzreturn)/4
   three_month_return=three_month_return.append(
       pd.Series(
       [
       n,
       p,
       wvreturn,
       xwreturn,
       xyreturn,
       yzreturn,
       avg
       ],
       index=collls),
       ignore_index=True
   )
three_month_return
```
